## clean
- extart number from price column and convert it to float
- replace none values with 'other' in "propert_type"column
- replace none values with 0 in 'review_scores_rating'column 


```python
df_seattle["price"] = df_seattle["price"].str.extract("(\d+\.\d+)").astype(float)
df_boston["price"] = df_boston["price"].str.extract("(\d+\.\d+)").astype(float)
 
```

```python
R = df_seattle['property_type'].isnull()
df_seattle.loc[R,'property_type'] = "Other"
R = df_boston['property_type'].isnull()
df_boston.loc[R,'property_type'] = "Other"
```

```python
R = df_seattle["review_scores_rating"].isnull()
df_seattle.loc[R,"review_scores_rating"] = 0
R = df_boston["review_scores_rating"].isnull()
df_boston.loc[R, "review_scores_rating"] = 0
```

```python
df_seattle = df_seattle.groupby("property_type",as_index=False).agg({
    "price":"mean","review_scores_rating":"mean",
    "id":"count"}) 
df_seattle.columns =["property_type" , "mean_price" ,"mean_rating","units_count"]

df_seattle
```

```python
df_boston = df_boston.groupby("property_type",as_index=False).agg({
    "price":"mean","review_scores_rating":"mean",
    "id":"count"}) 
df_boston.columns =["property_type" , "mean_price" ,"mean_rating","units_count"]

df_boston
```

```python
df_merge = df_seattle.merge(df_boston, on="property_type", suffixes=("_seattle","_boston"))
df_merge
```

```python
x = np.arange(df_merge.shape[0])
width = 0.35
fig, axs = plt.subplots(1, 2, figsize=(16,6) )
rectle1 = axs[0].bar(x - width/2, df_merge ["mean_price_seattle"], width, label="seattle")
rectle2 = axs[0].bar(x + width/2, df_merge ["mean_price_boston"], width, label="boston")

axs[0].set_ylabel("mean prices")
axs[0].set_title("mean prices seattle vs boston")
axs[0].set_xticks(x)
axs[0].set_xticklabels(df_merge["property_type"],rotation=90)

rectle1 = axs[1].bar(x - width/2, df_merge ["mean_rating_seattle"], width, label="seattle")
rectle2 = axs[1].bar(x + width/2, df_merge ["mean_rating_boston"], width, label="boston")

axs[1].set_ylabel("mean rtind")
axs[1].set_title("mean raiting seattle vs boston")
axs[1].set_xticks(x)
axs[1].set_xticklabels(df_merge["property_type"],rotation=90)



handles, labels = axs[0].get_legend_handles_labels()
fig.legend(handles, labels, loc="upper left")
fig.tight_layout()

```

```python
df_seattle =pd.read_csv("seattle/listings.csv") 
```

df_seattle.columns

```python
interest_columns = ["host_response_time",'host_acceptance_rate', 'host_is_superhost' , 'host_has_profile_pic',
   'host_identity_verified', 'neighbourhood_cleansed',  'property_type', 'room_type' , 'accommodates','bathrooms',    
    'bedrooms',"beds","price","number_of_reviews", "review_scores_rating"]
```

```python
df_seattle = df_seattle.loc [:,interest_columns ]
df_seattle 
```

```python
df_seattle.isnull().sum()
```

```python
df_seattle["host_response_time"].unique()
```

```python
R = df_seattle["host_response_time"].isnull()
df_seattle.loc[R,"host_response_time"] = "within an hour"

df_seattle["host_acceptance_rate"] = df_seattle["host_acceptance_rate"].str.extract(" (\d+) ").astype(float)
df_seattle["price"]=df_seattle["price"].str.extract("(\d+\.\d+)").astype(float)

R = df_seattle["host_acceptance_rate"].isnull()
df_seattle.loc[R,"host_acceptance_rate"] = 100

R = df_seattle["host_is_superhost"].isnull()
df_seattle.loc[R,"host_is_superhost"] =  'f'


R = df_seattle["host_has_profile_pic"].isnull()
df_seattle.loc[R,"host_has_profile_pic"] =  'f'

R = df_seattle["host_identity_verified"].isnull()
df_seattle.loc[R,"host_identity_verified"] ='f'
df_seattle.loc[:, ["host_is_superhost","host_has_profile_pic","host_identity_verified"]] = df_seattle.loc[:,["host_is_superhost","host_has_profile_pic","host_identity_verified"]].replace({"f":0,"t": 1})

response_dict = { 'within an hour': 1,'within a few hours': 2, 'within a day': 3,'a few days or more': 4}

df_seattle.loc[:,"host_response_time"] = df_seattle.loc[:,"host_response_time"].replace(response_dict)
R = df_seattle["property_type"].isnull()
df_seattle.loc[R,"property_type"] = "other"

R=df_seattle["review_scores_rating"].isnull()
df_seattle.loc[R, "review_scores_rating"] = 0
```

```python
R = (df_seattle['bathrooms'].isnull()) | (df_seattle['bedrooms'].isnull()) | (df_seattle['beds'].isnull())
df_seattle = df_seattle[~R]
```

```python
df_seattle.isnull().sum()
```

```python
df_seattle.sample(5)
```

```python
df_seattle = pd.get_dummies(df_seattle, columns=["neighbourhood_cleansed","property_type","room_type"])
df_seattle
```

```python
from sklearn.model_selection import train_test_split

x = df_seattle.drop("price", axis=1)

y = df_seattle["price"]

x_train, x_test, y_train,y_test = train_test_split(x, y, test_size=0.2, random_state = 42) 
```

```python
from sklearn.svm import SVR

reg = SVR(kernel="linear")
reg.fit(x_train, y_train)
```

```python
reg.score(x_train,y_train)
```

```python
reg.score(x_test,y_test)
```
